```python
import requests
import json
import pandas as pd

def geocode_address(address):
    try:
        parameters = {
            "key": "m2RWcdIQMSsY5tYulyaBK4j3fOojBh4Z",
            "location": address
        }
        response = requests.get("https://www.mapquestapi.com/geocoding/v1/address", params=parameters)
        data = json.loads(response.text)['results']
        if data:
            lat = data[0]['locations'][0]['latLng']['lat']
            lng = data[0]['locations'][0]['latLng']['lng']
            return lat, lng
    except Exception as e:
        print(f"Error geocoding address {address}: {e}")
    return None, None


batch_size = 1000  # Adjust the batch size as needed
for i in range(0, len(data), batch_size):
    batch_data = data.iloc[i:i+batch_size]
    for j, row in batch_data.iterrows():
        api_address = str(row['Accident_Road'])
        lat, lng = geocode_address(api_address)
        if lat is not None and lng is not None:
            data.at[j, 'lat'] = lat
            data.at[j, 'lng'] = lng
            
    
data.to_csv("accidentGeo.csv", index=False)
```
